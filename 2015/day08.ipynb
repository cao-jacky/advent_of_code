```python
import ast
import re
```

# part1

```python
day8_inputs = "inputs/day8.txt"

day8_raw_data = []

total_no_instructions = 0
with open(day8_inputs, 'r') as file:
    day8_data = file.readlines()
    total_no_instructions = len(day8_data)
```

```python
len_code = 0
len_in_memory = 0
for string in day8_data:
    len_code += len(string)
    len_in_memory += len(ast.literal_eval(string)) + 1

print("len_code", len_code)
print("len_in_memory", len_in_memory)

result = len_code - len_in_memory
print(result)
```

# part2

```python
# New encodings
encodings = {
    '"': r'\"',
    '\\': r'\\',
}
```

```python
total_chars = 0
for string in day8_data:
    encoded_string = ''
    for c in string:
        try:
            new_c = encodings[c]
        except:
            new_c = c
        encoded_string += new_c
    encoded_string = rf'"{encoded_string}"'
    total_chars += len(encoded_string)

print("total_chars", total_chars)
print("total_chars-len_code", total_chars-len_code)
```
