```python
from itertools import groupby
```

# part1

```python
max_process_loops = 40
starting_num = 1321131112

for i in range(max_process_loops):
    # Performing identical consecutive grouping
    starting_num_list = [int(x) for x in list(str(starting_num))]
    consec_groups = [list(y) for x, y in groupby(starting_num_list)]
    
    new_string = []
    for curr_group in consec_groups:
        curr_group_len = len(curr_group)
        new_string.append(curr_group_len)
        new_string.append(set(curr_group).pop())
    
    new_string = [f"{x}" for x in new_string]
    starting_num = ''.join(new_string)

print("length of result", len(starting_num))
```
