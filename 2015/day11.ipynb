```python
def over_slice(test_str, K):
    itr = iter(test_str)
    res = tuple(islice(itr, K))
    if len(res) == K:
        yield res    
    for ele in itr:
        res = res[1:] + (ele,)
        yield res
```

```python
def check_three_consec(string):
    consecutive_sequences = []
    current_sequence = string[0]  # Start with the first character
    
    for i in range(1, len(string)):
        # Check if the ASCII value of the current character is one more than the previous character
        if ord(string[i]) == ord(string[i - 1]) + 1:
            current_sequence += string[i]  # Append to current sequence
        else:
            # If the current sequence is longer than 1 character, save it
            if len(current_sequence) > 1:
                consecutive_sequences.append(current_sequence)
            current_sequence = string[i]  # Start a new sequence
    
    # Check if the last sequence is valid
    if len(current_sequence) > 1:
        consecutive_sequences.append(current_sequence)

    if len(consecutive_sequences[0]) > 3:
        return True
    else:
        return False
```

```python
def check_pairs(s):
    pairs = ["".join(ele) for ele in over_slice(s, 2)]
    num_pairs = 0
    for pair in pairs:
        if pair[0] == pair[1]:
            num_pairs += 1

    if num_pairs > 2:
        return True
    else:
        return False
```
