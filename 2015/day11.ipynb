```python
from itertools import islice
```

```python
def incr_chr(c):
    new_ord = (ord(c) - ord('a') + 1) % 26 + ord('a')
    wrapped = (new_ord == ord('a'))
    return chr(new_ord), wrapped

def incr_str(s):
    chars = list(s.lower())
    i = len(chars) - 1
    carry = True
    while i >= 0 and carry:
        chars[i], carry = incr_chr(chars[i])
        i -= 1
    if carry:
        chars.insert(0, 'a')
    return ''.join(chars)
```

```python
def over_slice(test_str, K):
    itr = iter(test_str)
    res = tuple(islice(itr, K))
    if len(res) == K:
        yield res    
    for ele in itr:
        res = res[1:] + (ele,)
        yield res
```

```python
def check_three_consec(string):
    sequence = ["".join(ele) for ele in over_slice(string, 3)]
    
    consecutive_sequences = []
    for seq in sequence:
        if (ord(seq[2]) - ord(seq[1])) == 1 and (ord(seq[1]) - ord(seq[0])) == 1:
            consecutive_sequences.append(seq)
        
    if len(consecutive_sequences) >= 1:
        return True
    else:
        return False
```

```python
def check_pairs(s):
    pairs_split = ["".join(ele) for ele in over_slice(s, 2)]
    pairs = []
    for pair in pairs_split:
        if pair[0] == pair[1]:
            pairs.append(pair)
    num_pairs = len(set(pairs))
    # print(set(pairs))
    if num_pairs >= 2:
        return True
    else:
        return False
```

```python
def check_invalid_chars(s):
    invalid_characters = ["i", "o", "l"]
    if any(x in s for x in invalid_characters):
        return False        
    else:
        return True
```

```python
initial_password = "vzbxkghb"
new_password = initial_password

# part 2
valid_passwords = []
num_passwords = 5
while len(valid_passwords) < num_passwords:
    new_password = incr_str(new_password)

    # Check second requirement (but use it as the first)
    req_2 = check_invalid_chars(new_password)
    
    # Check first requirement
    req_1 = check_three_consec(new_password)

    # Check third requirement
    req_3 = check_pairs(new_password)
    # print(new_password)
    
    if req_1 and req_2 and req_3:
        valid_passwords.append(new_password)
print(valid_passwords)
```
