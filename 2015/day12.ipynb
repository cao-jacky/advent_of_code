```python
import itertools
import json
import re
```

# part1

```python
day12_inputs = "inputs/day12.txt"

with open(day12_inputs, 'r') as file:
    day12_data = json.load(file)
```

```python
day12_numbers = re.findall(r'-?\d*\.?\d+', json.dumps(day12_data))
sum_all_nums = [sum(int(x) for x in day12_numbers)]
print(int(sum_all_nums[0]))
```

# part2

```python
def iterate_nested_dict(nested_dict):
    if "red" in nested_dict.values():
        return {}
    else:
        return nested_dict
```

```python
all_nums = str(json.loads(json.dumps(day12_data), object_hook=iterate_nested_dict))
day12_numbers = re.findall(r'-?\d*\.?\d+', all_nums)
sum_all_nums = [sum(int(x) for x in day12_numbers)]
print(int(sum_all_nums[0]))
```
