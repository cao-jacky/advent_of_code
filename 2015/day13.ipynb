```python
import itertools
import re
```

# part1

```python
day13_inputs = "inputs/day13.txt"
with open(day13_inputs, 'r') as file:
    day13_data = file.readlines()
```

```python
def neg_pos(change, points):
    if change == "gain":
        return points * 1
    elif change == "lose":
        return points * -1

curr_happiness = {}
happiness_rules = {} 
for row in day13_data:
    row_split = row.split(" ")
    
    primary_person = row_split[0]
    secondary_person = row_split[-1].strip().strip(".")
    
    happiness_points = int(re.findall(r'\d+', row)[0])
    happiness_change = neg_pos(row_split[2], happiness_points)

    happiness_rules.setdefault(primary_person, {})[secondary_person] = happiness_change
    curr_happiness.setdefault(primary_person, [0, 0])
    
people = list(curr_happiness.keys())
all_people_permutations = list(itertools.permutations(people))
```

```python
happiness_values = []
for perm in all_people_permutations:
    curr_perm_happiness = {}
    for i in range(len(perm)):
        person = perm[i]
        person_l = perm[i-1]
        if not i == len(perm)-1:
            person_r = perm[i+1]
        else:
            person_r = perm[0]
    
        person_l_happiness = happiness_rules[person][person_l]
        person_r_happiness = happiness_rules[person][person_r]
    
        curr_perm_happiness[person] = [person_l_happiness, person_r_happiness]
    happiness_list = [x for v in curr_perm_happiness.values() for x in v] 
    total_change_happiness = sum(happiness_list)
    happiness_values.append(total_change_happiness)
print(max(happiness_values))  
```
