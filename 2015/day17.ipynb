```python
from itertools import combinations
```

# part1

```python
day17_inputs = "inputs/day17.txt"
with open(day17_inputs, 'r') as file:
    day17_data = file.readlines()
```

```python
total_eggnog = 150
available_containers = [int(val.strip()) for val in day17_data]
```

```python
all_container_combos = [container for i in range(0, len(available_containers)) for container in combinations(available_containers, i)]

exact_combos = []
for container_combo in all_container_combos:
    vol_sum = sum(container_combo)
    if vol_sum == total_eggnog:
        exact_combos.append(container_combo)
print(len(exact_combos))
```

# part2

```python
# minimum number of containers, remove duplicate combos

min_num_containers = len(min(exact_combos, key=len))
min_num_lists = [combo for combo in exact_combos if len(combo) == min_num_containers]
print(len(min_num_lists))
```
