```python
import random
import re
```

# part1

```python
day19_inputs = "inputs/day19.txt"

with open(day19_inputs) as file:
    day19_data = file.readlines()
```

```python
replacements = {}
unique_molecules = [x for x in set([row.split("=>")[0].strip() for row in day19_data]) if 0 < len(x) <= 2]

for row in day19_data:
    if "=>" in row:
        split_row = row.split("=>")
        
        molecule_key = split_row[0].strip()    
        molecule_product = split_row[-1].strip()
        replacements.setdefault(molecule_key, []).append(molecule_product)

starting_molecule = str(day19_data[-1])
```

```python
generated_molecules = set()
for molecule, potential_replacements in replacements.items():
    molecule_locs = [m.start() for m in re.finditer(molecule, starting_molecule)]
    
    for replacement in potential_replacements:
        for loc in molecule_locs:
            new_molecule = starting_molecule[:loc] + replacement + starting_molecule[loc+len(molecule):]
            generated_molecules.add(new_molecule)

distinct_molecules = len(generated_molecules)
print(f"there are {distinct_molecules} distinct molecules")
```

# part2

```python
target_molecule = starting_molecule.strip()
basic_starting_molecule = "e"
```

```python
reversed_replacements = {}
for row in day19_data:
    if "=>" in row:
        split_row = row.split("=>")
        molecule_key = split_row[0].strip()    
        molecule_product = split_row[-1].strip()
        reversed_replacements.setdefault(molecule_product, molecule_key)
```

```python
curr_molecule = target_molecule
steps_taken = 0
reversed_replacements_shuffled = reversed_replacements
while curr_molecule != basic_starting_molecule:
    reversed_replacements_list = list(reversed_replacements_shuffled.items())
    random.shuffle(reversed_replacements_list)
    reversed_replacements_shuffled = dict(reversed_replacements_list)
    
    for molecule, replacement in reversed_replacements_shuffled.items():
        if molecule in curr_molecule:
            steps_taken += curr_molecule.count(molecule)
            curr_molecule = curr_molecule.replace(molecule, replacement)
            if curr_molecule == basic_starting_molecule:
                print(f"basic starting molecule of {basic_starting_molecule} obtained after {steps_taken} steps")
```
