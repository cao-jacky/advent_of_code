```python
import re
```

# part1

```python
day19_inputs = "inputs/day19.txt"

with open(day19_inputs) as file:
    day19_data = file.readlines()
```

```python
replacements = {}
unique_molecules = [x for x in set([row.split("=>")[0].strip() for row in day19_data]) if 0 < len(x) <= 2]

for row in day19_data:
    if "=>" in row:
        split_row = row.split("=>")
        
        molecule_key = split_row[0].strip()    
        molecule_product = split_row[-1].strip()
        replacements.setdefault(molecule_key, []).append(molecule_product)

starting_molecule = str(day19_data[-1])
```

```python
generated_molecules = set()
for molecule, potential_replacements in replacements.items():
    molecule_locs = [m.start() for m in re.finditer(molecule, starting_molecule)]
    
    for replacement in potential_replacements:
        for loc in molecule_locs:
            new_molecule = starting_molecule[:loc] + replacement + starting_molecule[loc+len(molecule):]
            generated_molecules.add(new_molecule)

distinct_molecules = len(generated_molecules)
print(f"there are {distinct_molecules} distinct molecules")
```

# part2

```python
target_molecule = starting_molecule
basic_starting_molecule = "e"
```

```python
def generate_molecules_recursive(initial_molecule, target_molecule, depths, replacements, max_depth=10, current_depth=0):
    if initial_molecule == target_molecule:
        depths.append(current_depth)
        return (True, current_depth)
    
    if current_depth >= max_depth:
        return (False, None)
    
    for molecule, potential_replacements in replacements.items():
        molecule_locs = [m.start() for m in re.finditer(molecule, initial_molecule)]
        
        for replacement in potential_replacements:
            for loc in molecule_locs: 
                new_molecule = (initial_molecule[:loc] + replacement + initial_molecule[loc+len(molecule):])
                
                found_molecule, steps = generate_molecules_recursive(new_molecule, target_molecule, depths, replacements, max_depth, current_depth + 1)
                
                if found_molecule:
                    return(True, steps)
                
    return (False, None)
```
