```python
replacements = {
    "H": ["HO", "OH"],
    "O": ["HH"]
}
starting_molecule = "HOHOHO"
```

```python
split_starting_molecule = list(starting_molecule)
print(split_starting_molecule)

generated_molecules = []
for molecule, potential_replacements in replacements.items():
    molecule_locs = [i for i, x in enumerate(split_starting_molecule) if x==molecule]
    print(molecule_locs)
    
    for replacement in potential_replacements:
        for loc in molecule_locs:
            new_molecule = list(starting_molecule)
            new_molecule[loc] = replacement
            
            mol_combined = ''.join(new_molecule)
            generated_molecules.append(mol_combined)
            print(loc, split_starting_molecule, molecule, replacement, mol_combined)

distinct_molecules = len(set(generated_molecules))
print(f"there are {distinct_molecules} distinct molecules from {starting_molecule}")
```
