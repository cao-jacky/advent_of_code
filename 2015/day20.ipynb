```python
def factors(n):
    return set(reduce(
        list.__add__,
        ([i, n//i] for i in range(1, int(n**0.5) + 1) if n % i == 0)))
```

```python
def prime_factorisation(input_number):
    test_prime = 2
    primes = []
    while test_prime * test_prime <= input_number:
        while input_number % test_prime == 0:
            primes.append(test_prime)
            input_number //= test_prime
        test_prime += 1
    if input_number > 1:
        primes.append(input_number)
    return primes
```

```python
def all_multiplications(input_list):
    all_container_combos = set([math.prod(multiple_set) for i in range(0, len(input_list)) for multiple_set in combinations(input_list, i)])
    
    print(all_container_combos)
```

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5)+1):
        if num % i == 0:
            return False
    return True
```
