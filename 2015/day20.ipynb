```python
import math
```

# part1

```python
puzzle_input = 34000000
```

```python
def get_divisors(house_num):
    elf = []
    limit = int(math.sqrt(house_num))
    for i in range(1, limit + 1):
        if house_num % i == 0:
            elf.append(i)
            other = house_num // i
            if other != i:
                elf.append(other)
    return elf
```

```python
house_num = 1
while True:
    total_presents = 10 * sum(get_divisors(house_num))
    if total_presents >= puzzle_input:
        print(house_num, total_presents)
        break
    house_num += 1
```

# part2

```python
house_num = 1
max_houses_per_elf = 50
while True:
    total_presents = 11 * sum(elf for elf in get_divisors(house_num) if house_num / elf <= max_houses_per_elf)
    if total_presents >= puzzle_input:
        print(house_num, total_presents)
        break
    house_num += 1
```
