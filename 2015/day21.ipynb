```python
day21_inputs = "inputs/day21.txt"
with open(day21_inputs, 'r') as file:
    day21_data = file.readlines()
day21_data_split = [el.strip().split(": ") for el in day21_data]
day21_data_flat = list(itertools.chain.from_iterable(day21_data_split))

boss = {day21_data_flat[i]: int(day21_data_flat[i + 1]) for i in range(0, len(day21_data_flat), 2)}

print(boss)
```

```python
gold_spent_wins = [] # part1
gold_spent_losses = [] # part2
for weapon_choice in list(shop["weapons"]):
    for armor_choice in list(shop["armor"]):
        for rings_choice in [container for i in range(0, 3) for container in combinations(list(shop["rings"]), i)]:
            player = {
                "Hit Points": 100,
                "Damage": 0,
                "Armor": 0,
                "Gold Spent":  0
            }
            
            player["Gold Spent"] = (shop["weapons"][weapon_choice]["cost"]
                                    + shop["armor"][armor_choice]["cost"]
                                    + sum([shop["rings"][ring_choice]["cost"] for ring_choice in rings_choice])
                                    )

            player["Damage"] = (shop["weapons"][weapon_choice]["damage"]
                                    + shop["armor"][armor_choice]["damage"]
                                    + sum([shop["rings"][ring_choice]["damage"] for ring_choice in rings_choice])
                                    )

            player["Armor"] = (shop["weapons"][weapon_choice]["armor"]
                                    + shop["armor"][armor_choice]["armor"]
                                    + sum([shop["rings"][ring_choice]["armor"] for ring_choice in rings_choice])
                                    )

            curr_boss_hit_points = boss["Hit Points"]
            curr_player_hit_points = player["Hit Points"]
            while True:
                # player attacks first
                player_damage = max(1, player["Damage"] - boss["Armor"])  # ensure at least 1 damage
                curr_boss_hit_points -= player_damage
                
                # boss attack
                boss_damage = max(1, boss["Damage"] - player["Armor"])  # ensure at least 1 damage
                curr_player_hit_points -= boss_damage
                
                if curr_boss_hit_points <= 0:
                    gold_spent_wins.append(player["Gold Spent"])
                    break
                if curr_player_hit_points <= 0:  # player died
                    gold_spent_losses.append(player["Gold Spent"])
                    break

print(min(gold_spent_wins))
print(max(gold_spent_losses))
```
