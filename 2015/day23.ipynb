```python
day23_inputs = "inputs/day23.txt"
with open(day23_inputs, 'r') as file:
    day23_data = file.readlines()
day23_data = [el.strip() for el in day23_data]
```

```python
def instructions_loop(instructions, registers):
    i = 0
    while 0 <= i < len(instructions):
        instruction = instructions[i]
        primary_instruction = instruction[0]
        referenced_variable = instruction[1].strip(",")
                
        if "hlf" == primary_instruction:
            registers[referenced_variable] //= 2
            i += 1
        elif "tpl" == primary_instruction:
            registers[referenced_variable] *= 3 
            i += 1
        elif "inc" == primary_instruction:
            registers[referenced_variable] += 1
            i += 1
        elif "jmp" == primary_instruction:
            offset = instruction[1]
            i += int(offset)
        elif "jie" == primary_instruction:
            if registers[referenced_variable] % 2 == 0:
                offset = instruction[2]
                i += int(offset)
            else:
                i +=1
        elif "jio" == primary_instruction:
            if registers[referenced_variable] == 1:
                offset = instruction[2]
                i += int(offset)
            else:
                i +=1
    return registers
```

```python
instructions = [line.split() for line in day23_data]

registers = {"a": 0, "b": 0}
part1 = instructions_loop(instructions, registers)
print("part1", part1)

registers = {"a": 1, "b": 0}
part2 = instructions_loop(instructions, registers)
print("part2", part2)
```
