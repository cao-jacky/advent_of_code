```python
import numpy as np
import re
```

```python
day25_inputs = "inputs/day25.txt"
with open(day25_inputs, 'r') as file:
    day25_data = file.readlines()
target_row, target_col = [int(val) for val in re.findall(r'\d+', day25_data[0])]
print(target_row, target_col)
```

```python
# if I create a large enough array to calculate the codes for the portion of the
# codes that we want to consider, the correct answer for the specified position
# should exist, and can be easily indexed out 

initial_val = 20151125

prev_code = 0
rows = target_col * 2
cols = rows

for line in range(2, rows + cols + 1):
    start_row = min(line - 1, rows)
    start_col = line - start_row
    while start_row >= 1 and start_col <= cols:
        if prev_code == 0:
            next_code = initial_val
        else:
            next_code = (prev_code * 252533) % 33554393
        if start_row == target_row and start_col == target_col:
            print(f"part1: {next_code}")
            break
        prev_code = next_code
        start_row -= 1
        start_col += 1
    else:
        continue
    break
```
