```python
import ast
import re

import pandas as pd
import numpy as np

from itertools import permutations
```

# part1

```python
day9_inputs = "inputs/day9.txt"

with open(day9_inputs, 'r') as file:
    day9_data = file.readlines()
```

```python
print(day9_data)
```

```python
potential_paths = []
for travel_path in day9_data:
    pp_dict = {}
    
    parts = re.findall(r'\w+', travel_path.strip())
    pp_dict["loc1"] = parts[0]
    pp_dict["loc2"] = parts[2]
    pp_dict["distance"] = parts[-1]
    
    potential_paths.append(pp_dict)

pp_df = pd.DataFrame(potential_paths)
```

```python
all_locs = np.union1d(pp_df["loc1"].unique(), pp_df["loc2"].unique())
```

```python
# All potential permutations (before checking if they are valid)
all_pot_perms = list(permutations(all_locs))
```

```python
all_dists = []
for perm in all_pot_perms:
    prev_loc = ""
    total_dist = 0
    
    for curr_loc in perm:
        if prev_loc:
            locs = {prev_loc, curr_loc}
    
            # Apply pairwise condition to filter out the rows
            filtered_path = pp_df[
                ((pp_df['loc1'] == prev_loc) & (pp_df['loc2'] == curr_loc)) |
                ((pp_df['loc1'] == curr_loc) & (pp_df['loc2'] == prev_loc))
            ]
            total_dist += int(filtered_path["distance"].iloc[0])
        prev_loc = curr_loc
    all_dists.append(total_dist)
```

```python
print("shortest route is", min(all_dists))
```

# part2

```python
print("longest route is", max(all_dists))
```
