```python
import numpy as np
import re
```

```python
day01_inputs = "inputs/day01.txt"
with open(day01_inputs, 'r') as file:
    day01_data = file.readlines()
```

```python
dir_dict = {
    "1": {
        "dir": "North",
        "L": -2,
        "R": 2,
        "data_index": 1
    },
    "-1": {
        "dir": "South",
        "L": 2,
        "R": -2,
        "data_index": 1
    },
    "2": {
        "dir": "East",
        "L": 1,
        "R": -1,
        "data_index": 0
    },
    "-2": {
        "dir": "West",
        "L": -1,
        "R": 1,
        "data_index": 0
    }
}
```

```python
# dir_instr = "R8, R4, R4, R8"
dir_instr = day01_data[0].strip()
dir_instr_split = dir_instr.split(", ")
```

```python
curr_loc = [0,0]
curr_dir = 1 # directions NESW, are 1, 2, -1, -2

visited_locs = []
for instr in dir_instr_split:    
    instr_dir = re.findall("[a-zA-Z]+", instr)[0]
    instr_dist = int(re.findall("[0-9]+", instr)[0])
    
    prev_loc = curr_loc.copy()
    
    new_dir = dir_dict[str(curr_dir)][instr_dir] 
    new_card_dir = dir_dict[str(new_dir)]["dir"]
    
    axis = dir_dict[str(new_dir)]["data_index"]
    if new_dir in [1, 2]: # North or East
        curr_loc[axis] += instr_dist
    else: # South or West
        curr_loc[axis] -= instr_dist
        
    dx = curr_loc[0] - prev_loc[0]
    dy = curr_loc[1] - prev_loc[1]
    
    dx_sign = int(np.sign(dx))
    dy_sign = int(np.sign(dy))
    
    intermediate_point = prev_loc.copy()
    last_point = intermediate_point.copy()
    calc_point = True
    while calc_point:
        intermediate_point = [intermediate_point[0] + dx_sign, intermediate_point[1] + dy_sign]
        if tuple(intermediate_point) in visited_locs:
            blocks_away = abs(intermediate_point[0]) + abs(intermediate_point[1])
            print(f"part2: {intermediate_point} has already been visited, this location is {blocks_away} blocks away")
        if intermediate_point == curr_loc:
            calc_point = False
        last_point = intermediate_point
        visited_locs.append(tuple(intermediate_point))
    curr_dir = new_dir
    
blocks_away = abs(curr_loc[0]) + abs(curr_loc[1])
print(f"part1: location is {blocks_away} blocks away")
```
