```python
import polars as pl
```

```python
day03_inputs = "inputs/day03.txt"
```

# part1

```python
df = (
    pl.read_csv(
        day03_inputs,
        has_header=False,
        new_columns=["data"],
    )
    
    .select(
        pl.col("data")
        .str.split(" ")
        .list.eval(pl.element().filter(pl.element() != ""))
        .list.eval(pl.element().cast(pl.Int64))
        .list.sort() # sort into ascending order
        .list.to_struct(fields=["side1", "side2", "side3"])
        .alias("data")
    )
    .unnest("data")
    
    .filter(pl.col("side1") + pl.col("side2") > pl.col("side3"))
)

print(f"{len(df)} possible triangles")
```

# part2

```python
df = (
    pl.read_csv(
        day03_inputs,
        has_header=False,
        new_columns=["data"],
    )
    
    .select(
        pl.col("data")
        .str.split(" ")
        .list.eval(pl.element().filter(pl.element() != ""))
        .list.eval(pl.element().cast(pl.Int64))
        .list.to_struct(fields=["side1", "side2", "side3"])
        .alias("data")
    )
    .unnest("data")
)

arr = df.to_numpy().T.reshape(-1, 3).T
arr.sort(axis=0)
df = (
    pl.DataFrame(arr, schema=["side1", "side2", "side3"])
    .filter(pl.col("side1") + pl.col("side2") > pl.col("side3"))
)
print(f"{len(df)} possible triangles")
```
