```python
import re

from collections import Counter
```

```python
day04_inputs = "inputs/day04.txt"
with open(day04_inputs, 'r') as file:
    day04_data = file.readlines()
day04_data = [line.strip() for line in day04_data]
```

# part1

```python
valid_rooms_sector_ids = []
valid_encrypted_rooms = []
for room_name in day04_data:
    irn_split = re.match(r"^(.*?)\[(.*?)\]$", room_name)
    room_sector = irn_split.group(1)

    sector_id = int(re.findall(r'\d+', room_sector)[0])
    encrypted_name = room_sector.replace(f"-{sector_id}", "")
    checksum = list(irn_split.group(2))

    count_characters = Counter(encrypted_name)
    sorted_cc = {k: v for k, v in sorted(count_characters.items(), key=lambda t: (-t[1], t[0])) if k != "-"}

    top_five_chars = [i for i in sorted_cc.keys()][:5]

    if Counter(checksum) == Counter(top_five_chars):
        valid_rooms_sector_ids.append(sector_id)
        valid_encrypted_rooms.append(room_sector)

print(f"part1: {sum(valid_rooms_sector_ids)}")
```

# part2

```python
valid_encrypted_rooms = valid_encrypted_rooms
for room_name in valid_encrypted_rooms:
    sector_id = int(re.findall(r'\d+', room_name)[0])
    encrypted_name = room_name.replace(f"-{sector_id}", "")
    
    decrypted_name = "".join(
        " " if c == "-" else
        chr((ord(c) - ord('a') + sector_id) % 26 + ord('a'))
        for c in encrypted_name
    )
    if "north" in decrypted_name:
        print(f'part2: North Pole objects are stored in room with name "{decrypted_name}", and sector id {sector_id}')    
```
