```python
import hashlib
```

```python
puzzle_input = "reyedfim"
```

# part1

```python
door_id = puzzle_input
integer_index = 0
password = []

while len(password) < 8:
    hash_string = f"{door_id}{integer_index}"
    hash_hex = hashlib.md5(hash_string.encode()).hexdigest()
    
    if hash_hex.startswith("00000"):
        password_char = hash_hex[5]
        password.append(password_char)
        print(password)
    integer_index += 1

final_password = "".join(password)
print(f'part1: password to door "{door_id}" is {final_password}')
```

# part2

```python
door_id = puzzle_input
integer_index = 0
password = ["_"] * 8

while "_" in password:
    hash_string = f"{door_id}{integer_index}"
    hash_hex = hashlib.md5(hash_string.encode()).hexdigest()
    
    if hash_hex.startswith("00000"):
        char_position = hash_hex[5]
        if char_position.isdigit() and int(char_position) < 8:
            password_position = int(char_position)
            if password[password_position] == "_":
                password[password_position] = hash_hex[6]
                print(password)    
    integer_index += 1

final_password = "".join(password)
print(f'part2: password to door "{door_id}" is {final_password}')
```
