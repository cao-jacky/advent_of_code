```python
import polars as pl
```

```python
day06_inputs = "inputs/day06.txt"
with open(day06_inputs, 'r') as file:
    day06_data = file.readlines()
```

```python
df = (
    pl.read_csv(
        day06_inputs,
        has_header=False,
        new_columns=["data"],
    )
    
    .select(
        pl.col("data")
        .str.split("")
        .list.to_struct(upper_bound=len(day06_data[0].strip()))
    )
    .unnest("data")
    
    .select(
        pl.all()
        .value_counts(sort=True)
    )
)

print(df)

part1_message = "".join([df.select(df.columns[i]).to_series()[0][f"field_{i}"] for i in range(df.width)])
print(f'part1: the message is "{part1_message}"')

part2_message = "".join([df.select(df.columns[i]).to_series()[-1][f"field_{i}"] for i in range(df.width)])
print(f'part2: the message is "{part2_message}"')
```
