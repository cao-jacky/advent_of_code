```python
import re
```

```python
day07_inputs = "inputs/day07.txt"
with open(day07_inputs, 'r') as file:
    day07_data = file.readlines()
day07_data = [line.strip() for line in day07_data]
```

```python
def find_palindromes(test_string, problem_part, string_len=4):
    palindromes = []
    num_chars = len(test_string)
    for char in range(num_chars):
        for next_char in range(char+string_len, num_chars+1):
            search_string = test_string[char:next_char]
            if search_string == search_string[::-1]:
                # check if same character at positions 0 and 2, if same, then it is an invalid palindrome
                if len(search_string) == string_len: 
                    if problem_part == "part1":
                        if not search_string[0] == search_string[2]:
                            palindromes.append(search_string)
                    elif problem_part == "part2":
                        if not search_string[0] == search_string[1]:
                            palindromes.append(search_string)
    return palindromes
```

# part1

```python
allowed_ips = []

for input_ip in day07_data:
    skip_ip = False
    hypernet_sequences = re.findall(r'\[(.*?)\]', input_ip)
    for h_seq in hypernet_sequences:
        # if any palindrome in any hypernet sequence, skip current IP
        if find_palindromes(h_seq, "part1", 4):
            skip_ip = True
            break
    
    if skip_ip:
        continue
        
    non_hypernet_data = re.split(r'\[.*?\]', input_ip)
    for search_seq in non_hypernet_data:
        search_palindromes = find_palindromes(search_seq, "part1", 4)
        if search_palindromes:
            allowed_ips.append(input_ip)
    
allowed_ips = set(allowed_ips)
print(f"there are {len(allowed_ips)} IPs which support TLS")
```

# part2

```python
allowed_ips = []

for input_ip in day07_data:    
    non_hypernet_data = re.split(r'\[.*?\]', input_ip)
    aba_seqs = []
    required_bab_seqs = []
    
    for search_seq in non_hypernet_data:
        search_palindromes = find_palindromes(search_seq, "part2", 3)
        if search_palindromes:
            aba_seqs.extend(search_palindromes)
            required_bab_seqs.extend(f"{seq[1]}{seq[0]}{seq[1]}" for seq in search_palindromes)
            
    hypernet_sequences = re.findall(r'\[(.*?)\]', input_ip)
    found_bab = False
    for h_seq in hypernet_sequences:
        for req_bab_seq in required_bab_seqs:
            if req_bab_seq in h_seq:
                found_bab = True
                break
        if found_bab:
            break
    if found_bab:
        allowed_ips.append(input_ip)
            
allowed_ips = set(allowed_ips)
print(f"there are {len(allowed_ips)} IPs which support SSL")
```
